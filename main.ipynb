### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nbconvert

import warnings
warnings.filterwarnings('ignore')
```

### Data Cleaning

In this section, we will remove unwanted variables or values from the dataset. Some things to do are removing unnecessary rows/columns and also reformatting our data so we can get data that we want.

```python
#Importing data

data = pd.read_csv("data/Amazon_Products.csv", low_memory=False)
data.head()
```

```python
#Look at the number of row and columns
data.shape
```

```python
data.columns
```

```python
#Check the missing value
data.isna().sum()
```

```python
#Check precentage of missing value from every columns

for col in data.columns:
    pct_missing = data[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
#Removing columns with missing value more than 90%

perc = 90.0 
max_nan =  int(((100 - perc)/100) * data.shape[1])
data.dropna(axis=1, thresh=max_nan,inplace=True)
```

```python
#Dimension after progress
data.shape
```

```python
#Columns
data.columns
```

```python
data.dtypes
```

#### Transform data

It is shown that all the data types are categorical. We need to transform the data to get what we need.

```python
#Rename and change dtype of number of review column

#rename column name
data.rename(columns={"number_of_reviews" : "total_reviews"}, inplace=True) 

#convert to numeric type
data["total_reviews"] = pd.to_numeric(data.total_reviews, errors='coerce')

#convert to float
data["total_reviews"] = data["total_reviews"].astype("float")
```

```python
#Price must be numeric value

pattern_price = r'(\d+.\d\d)'
data["product_price"] = data["price"].str.replace(r',','.').str.extract(pattern_price).astype('float')
```

```python
#Stock must be numeric value

pattern_stock = r'(\d+)'
data["stock"] = data["number_available_in_stock"].str.extract(pattern_stock).astype("float")
```

```python
#Rating must be numeric value

pattern_rate = r'(\d.\d)'
data["rating"]= data["average_review_rating"].str.extract(pattern_rate).astype("float")
```

```python
#Split main category into multiple sub-category

cut = data["amazon_category_and_sub_category"].str.split(">", expand=True)
data['main_category'] = cut.get(0)
data['sub_category_1'] = cut.get(1)
data['sub_category_2'] = cut.get(2)
data['sub_category_3'] = cut.get(3)
```

```python
#New dataframe containing only the columns that we need

amazon_product = data[['product_name','manufacturer','product_price',
                       'stock','total_reviews','rating','main_category',
                       'sub_category_1','sub_category_2','sub_category_3']]
amazon_product
```

```python
amazon_product.info()
```

Now we have Amazon product data with 6 categorical columns and 4 numerical columns.

```python
#Saving data to csv file
amazon_product.to_csv("Amazon_Product_2020.csv", index=False)
```

## Data Visualization

Part A - Top 10 Manufacturer

Our further analysis based on rating so we will drop missing values from the columns.

```python
df = amazon_product.dropna(subset=["rating"])
```

```python
#Find top 10 manufacturer
mfr_value = df["manufacturer"].value_counts().keys().tolist()[:10]
mfr_count = df["manufacturer"].value_counts().tolist()[:10]
```

```python
#Plotting
fig = plt.figure(figsize=(20,10))

mfr_plot = sns.barplot(mfr_value, mfr_count, palette="pastel", label=mfr_count)
for i in mfr_plot.containers:
    mfr_plot.bar_label(i)

plt.xlabel("\nManufacturer",fontsize=15)
plt.ylabel("Count", fontsize=15, )
plt.title("Top 10 Manufacturer", fontsize=25)
plt.xticks(fontsize=12,rotation=45)

plt.show()
```

Based on the bar chart, it can be seen that LEGO is the most popular manufacturer.

```python
#Creating dataframe of top 10 manufacturer
mfr = []
for i in mfr_value:
    x = df[df["manufacturer"] == i]
    mfr.append(x)
top_mfr = pd.concat(mfr)
top_mfr.head()
```

```python
#Average rating based on manufacturer
avg_rate = top_mfr.groupby("manufacturer")[["rating"]].mean()
avg_rate.sort_values(by=["rating"],ascending=False)
```

From the table, we know that even though LEGO is the most popular manufacturer but MyTinyWorld has the highest rating.

Part B - Popular Category of Top 10 manufacturer

```python
#Find Popular Category
cat_value = top_mfr["main_category"].value_counts().keys().tolist()[:5]
cat_count = top_mfr["main_category"].value_counts().tolist()[:5]
```

```python
#Percentage of top 5 main category
total = sum(cat_count)
perc_cat=[]
for i in cat_count:
    perc = round((i/total)*100)
    perc_cat.append(perc)
```

```python
#Plotting

colors = sns.color_palette('pastel')[0:5]
plt.figure(figsize=(7,7))
plt.pie(perc_cat, labels=cat_value, colors=colors, autopct="%.0f%%")
plt.title("Top 5 Main Category")
plt.show()
```

As shown in the chart, category Characters & Brands is the popular main category and more likely to become bestseller category.

```python
#Filtering dataframe of top 5 main category

top_cat = top_mfr[top_mfr.main_category.isin(cat_value)]
```

```python
#Average rating based on main category
avg_rate_category = top_cat.groupby("main_category")[["rating"]].mean()
avg_rate_category.sort_values(by=["rating"],ascending=False)
```

From the tabel we know that Dolls & Accesories has the highest rating even though Characters & Brands is the most popular category.
